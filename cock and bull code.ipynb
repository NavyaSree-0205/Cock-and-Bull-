```python
import random

# Display welcome message
print("Welcome to the Fun and Tricky game 'Cock and Bull'!")

# Ask for the player's name
player_name = input("Please enter your name: ")

# Ask for the number of tries
tries = int(input("Enter number of tries: "))

# Function to generate a secret 4-digit number with unique digits
def generate_secret_number():
    digits = list(range(10))
    random.shuffle(digits)
    secret_number = ''.join(map(str, digits[:4]))
    return secret_number

# Generate the secret number
secret_number = generate_secret_number()

print(secret_number)  # For demonstration purposes only

# Function to get a valid guess from the player
def get_guess():
    while True:
        guess = input("Enter your guess: ")
        if not guess.isdigit():
            print("Enter only digits.")
            continue
        if len(guess) == 4 and len(set(guess)) == 4:
            return guess
        print("Invalid guess. Please enter a 4-digit number with unique digits.")

# Function to calculate cows and bulls
def calculate_cows_and_bulls(secret_number, guess):
    bulls = sum(s == g for s, g in zip(secret_number, guess))
    cows = sum(min(secret_number.count(d), guess.count(d)) for d in set(guess)) - bulls
    return cows, bulls

# Function to play the game
def play_game():
    secret_number = generate_secret_number()
    attempts = 0
    score = 0

    while attempts < tries:
        guess = get_guess()
        attempts += 1
        cows, bulls = calculate_cows_and_bulls(secret_number, guess)
        print(f"{bulls} Bulls, {cows} Cows")

        score += 1

        if bulls == 4:
            print(f"Congratulations! You've guessed the secret number {secret_number} in {attempts} attempts.")
            break
    else:
        print(f"You ran out of tries. The secret number was {secret_number}.")

    return score

# Play the game and display the score
score = play_game()
print(f"{player_name}, your score is {score}.")

# Function to filter possible numbers based on feedback
def filter_possible_numbers(possible_numbers, guess, cows, bulls):
    def matches_feedback(number):
        test_cows, test_bulls = calculate_cows_and_bulls(number, guess)
        return test_cows == cows and test_bulls == bulls

    return [number for number in possible_numbers if matches_feedback(number)]

# Function for the computer to guess the secret number
def computer_guess_game():
    possible_numbers = [f"{i}{j}{k}{l}" for i in range(10) for j in range(10) for k in range(10) for l in range(10)
                        if len({i, j, k, l}) == 4]
    attempts = 0

    while True:
        guess = possible_numbers[0]
        attempts += 1
        print(f"Computer's guess: {guess}")
        cows, bulls = calculate_cows_and_bulls(secret_number, guess)
        print(f"{bulls} Bulls, {cows} Cows")

        if bulls == 4:
            print(f"Computer guessed the secret number {secret_number} in {attempts} attempts.")
            break

        possible_numbers = filter_possible_numbers(possible_numbers, guess, cows, bulls)

# Run the computer guess game
computer_guess_game()
```
